Again working on heatmaps, but this time it's including a lot of different sizes of grids.

```python
import pandas as pd  # Data handling
import matplotlib.pyplot as plt # visualization
from scipy.integrate import odeint # numerical integration
import numpy as np
import seaborn as sns
```

```python
# read in data
small_grid = np.loadtxt("/Users/gracerojo/Downloads/DiffEqThings/metrics.csv", delimiter=",")
big_grid = np.loadtxt("/Users/gracerojo/Downloads/DiffEqThings/metrics_grid.csv", delimiter=",")
smaller_grid = np.loadtxt("/Users/gracerojo/Downloads/DiffEqThings/metrics_grid_small_grace.csv", delimiter=",")
even_smaller_grid = np.loadtxt("/Users/gracerojo/Downloads/metrics_grid.csv", delimiter=",")
new_smaller_grid = np.loadtxt("/Users/gracerojo/Downloads/metrics_grid_new.csv", delimiter=",")
```

```python
# double checking that big grid actually has the right data in it
count = 0
for entry in big_grid:
    if entry != 0.0:
        count += 1
    else:
        pass
print(count)
```

```python
# smaller grid heatmap
a_min = 0.0
a_max = 0.5
b_min = 1.0
b_max = 5.0
N = 5
grid = np.zeros((N,N))
count = 0
for i,a in enumerate(np.linspace(a_min,a_max,N)):     # a will be y axis in grid
    for j,b in enumerate(np.linspace(b_min,b_max,N)): # b will be x axis in grid
        grid[i,j] = small_grid[count]
        count += 1

colormap = sns.color_palette("Blues_r", as_cmap=True)
plt.figure(figsize=(8,6))
sns.heatmap(np.log(np.flip(grid, axis = 0)),cmap = colormap)
plt.title(r"log of euclidean metric results for $\ddot{x} = -x -ax^b$")
plt.xlabel("b")
plt.ylabel('a')
plt.xticks(ticks = np.arange(N)+0.5, labels = np.around(np.linspace(b_min,b_max,N),2))
plt.yticks(ticks = np.arange(5)+0.5, labels = np.around(np.linspace(a_max,a_min,N),2))
plt.show()
```
